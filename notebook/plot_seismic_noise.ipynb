```python
import numpy as np
import matplotlib.pyplot as plt

seismic_noise_90_x_data = np.loadtxt("data/X_90_DISP.txt")
seismic_noise_90_y_data = np.loadtxt("data/Y_90_DISP.txt")
seismic_noise_90_z_data = np.loadtxt("data/Z_90_DISP.txt")

f_x, x = seismic_noise_90_x_data[:, 0], seismic_noise_90_x_data[:, 1]
f_y, y = seismic_noise_90_y_data[:, 0], seismic_noise_90_y_data[:, 1]
f_z, z = seismic_noise_90_z_data[:, 0], seismic_noise_90_z_data[:, 1]
```

```python
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(10, 5))
plt.title("KAGRA seismic noise at the $90^\mathrm{th}$ percentile.")
plt.loglog(f_x, x, label="X direction")
plt.loglog(f_y, y, label="Y direction")
plt.loglog(f_z, z, label="Z direction")
plt.ylabel("Amplitude spectral density ($\mu\mathrm{m}/\sqrt{\mathrm{Hz}}$)")
plt.xlabel("Frequency (Hz)")
plt.legend(loc=0)
plt.grid(which="both")
plt.savefig("kagra_seismic_noise_90th_percentile")
```
