```python
import control
import numpy as np
import matplotlib.pyplot as plt

s = control.tf("s")
w = np.logspace(-1, 5, 1000)

G = 2 * (s/1+1) / (s/10+1)**2 * (s**2/100**2 + 1/(100*30)*s + 1) / (s**2/1000**2 + 1/(1000*300)*s + 1)

plt.figure(figsize=(12, 6))
plt.rcParams["font.size"] = 14
plt.loglog(w, abs(G(1j*w)), label=r"$G(s) = 2\frac{(s/1 + 1)}{(s/10 + 1)^2}\frac{(s^2/100^2 + \frac{1}{100\times 30}s + 1)}{(s^2/1000^2 + \frac{1}{1000\times300}s + 1)}$")
plt.legend(loc=0)
plt.grid(which="both")
plt.ylabel(r"Magnitude $|G(j\omega)|$")
plt.xlabel("Frequency (rad/s)")

plt.annotate('DC gain of 2', xy=(0.1, 3), xytext=(0.1, 100), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate('simple zero at $\omega=1$', xy=(1, 2), xytext=(1, 0.1), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate(r'$\propto \omega^{0+1}=\omega^1$', xy=(3, 4), xytext=(5, 1), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate('Two simple poles at $\omega=10$', xy=(10, 20), xytext=(10, 100), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate(r'$\propto \omega^{1-2}=\omega^{-1}$', xy=(40, 5), xytext=(50, 20), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate('Anti-resonance $\omega=100$', xy=(120, 0.1), xytext=(300, 0.1), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate('Resonance $\omega=1000$', xy=(1200, 4000), xytext=(3000, 4000), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.annotate(r'$\propto \omega^{-1+2-2}=\omega^{-1}$', xy=(10000, 3), xytext=(12000, 10), 
            arrowprops=dict(facecolor='black', shrink=0.),
            )

plt.title("Bode magnitude plot")

plt.savefig("../figures/example_transfer_function_plot.png", bbox_inches="tight")
```
